```python
from sklearn.datasets import load_iris

Iris = load_iris()
```

```python
import numpy as np
import pandas as pd

Iris_data = pd.DataFrame(data = Iris['data'], columns = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
```

```python
Iris_data
```

```python
Iris_target = pd.DataFrame(data = Iris['target'], columns = ['target'])
```

```python
Iris_target
```

```python
Iris_data.shape
```

```python
Iris_target.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from keras.callbacks import EarlyStopping
```

```python
y = np_utils.to_categorical(Iris_target)
y.shape
```

```python
model = Sequential()
model.add(Dense(16, input_dim = 4, activation = 'relu'))
model.add(Dense(3, activation = 'softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
hist = model.fit(Iris_data, y, epochs = 150, batch_size = 1)
```

```python
print("\n Accuracy: %.4f" % (model.evaluate(Iris_data, y)[1]))
```
